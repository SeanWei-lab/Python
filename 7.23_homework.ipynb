```python
#1题
class Rectangle(object):
    def __init__(self,width=1,height=2):
        self.width=width
        self.height=height
    def getArea(self):
        area=self.width*self.height
        return area
    def getPerimeter(self):
        Perimeter=self.width*2+self.height*2
        return Perimeter
if __name__=='__main__':
    for i in range(2):
        width,height=map(int,input("输入宽和高>>>").split())
        rec=Rectangle(width,height)
        print(rec.width,rec.height,rec.getArea(),rec.getPerimeter())
```

```python
#2题
class Account(object):
    def __init__(self,id,balance=100,annuallnterestRate=0):
        self._id=id
        self._balance=balance
        self._annuallnterestRate=annuallnterestRate
    def getMonthlyInterestRate(self):
        self.MonthlyInterestRate=self._annuallnterestRate/100/12
        return self.MonthlyInterestRate
    def getMonthlyInterest(self):
        MonthlyInterest=self.MonthlyInterestRate*self._balance
        return MonthlyInterest
    def withdraw(self,go_):
        self._balance=self._balance-go_
        return self._balance
    def deposit(self,come_):
        self._balance=self._balance+come_
        return self._balance
id=input('请输入id：')
blance=int(input('请输入起始额：'))
annuallnterestRate=float(input('请输入年利率：'))
account = Account(id,blance,annuallnterestRate)
print(account._id,account._balance,account.getMonthlyInterestRate(),account.getMonthlyInterest())
account.withdraw(2500)
print(account._balance)
account.deposit(3000)
print(account._balance)
```

```python
#3题
class Fan(object):
    def __init__(self,speed=1,on=False,radius=5,color='blue'):
        self.SLOW=1
        self.MEDIUM=2
        self.FAST=3
        self.speed=speed
        self.on=on
        self.radius=radius
        self.color=color
    def getSpecail(self):
        speed1='close'
        if self.speed=='1':
            speed1='SLOW'
        elif self.speed=='2':
            speed1='MEDIUM'
        elif self.speed=='3':
            speed1='FAST'
        on1=self.on
        radius1=self.radius
        color1=self.color
        return speed1,radius1,color1,on1
speed=input('输入速度5：')
radius=input('输入半径：')
color=input('输入颜色：')
on=input('输入是否打开：')
fan = Fan(speed,radius,color,on)
print(fan.getSpecail())
speed=input('输入速度：')
radius=input('输入半径：')
color=input('输入颜色：')
on=input('输入是否打开：')
fan = Fan(speed,radius,color,on)
print(fan.getSpecail())
```

```python
#4题
class RegularPolygon(object):
    def __init__(self,n=3,side=1,x=0,y=0):
        self.n=n
        self.side=side
        self.x=x
        self.y=y
        self.perimeter=0
        self.area=0
    def getPerimeter(self):
        self.perimeter=self.n*self.side
        return self.perimeter
    def getArea(self):
        import math
        self.area=(self.n*self.side**2)/(4*math.tan(math.pi/self.n))
        return self.area
regularPolygon=RegularPolygon()
print('面积为：'+str(regularPolygon.getArea()))
print('周长为：'+str(regularPolygon.getPerimeter()))
regularPolygon1=RegularPolygon(6,4)
print('面积为：'+str(regularPolygon1.getArea()))
print('周长为：'+str(regularPolygon1.getPerimeter()))
regularPolygon2=RegularPolygon(10,4,5.6,7.8)
print('面积为：'+str(regularPolygon2.getArea()))
print('周长为：'+str(regularPolygon2.getPerimeter()))
```

```python
#5题
class LinearEquation(object):   
    def __init__(self,a,b,c,d,e,f):
        self.__a=a
        self.__b=b
        self.__c=c
        self.__d=d
        self.__e=e
        self.__f=f           
    def set_a(self,a):
        self.__a=a
    def get_a(self):
        return self.__a
    def set_b(self,b):
        self.__b=b
    def get_b(self):
        return self.__b
    def set_c(self,c):
        self.__c=c
    def get_c(self):
        return self.__c
    def set_d(self,d):
        self.__d=d
    def get_d(self):
        return self.__d
    def set_e(self,e):
        self.__e=e
    def get_e(self):
        return self.__e
    def set_f(self,f):
        self.__f=f
    def get_f(self):
        return self.__f
    def isSolvable(self):
        if (self.__a*self.__d)-(self.__c*self.__b)!=0:
            return True
        else:
            return print('此方程无解')
    def getX(self):
        s=(self.__a*self.__d)-(self.__b*self.__c)
        x=(self.__e*self.__d)-(self.__b*self.__f)/s
        print('X的值为：%.2f'% x)
    def getY(self):
        s=(self.__a*self.__d)-(self.__b*self.__c)
        y=(self.__a*self.__f)-(self.__e*self.__c)/s
        print('Y的值为：%.2f'% y)
if __name__=="__main__":
    a=int(input('a的值是：'))
    b=int(input('b的值是：'))
    c=int(input('c的值是：'))
    d=int(input('d的值是：'))
    e=int(input('e的值是：'))
    f=int(input('f的值是：'))
    l=LinearEquation(a,b,c,d,e,f)
    l.isSolvable()
    l.getX()
    l.getY()
```

```python
#6题
class Point(object):
    x =0
    y= 0
    # 定义构造方法
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

class Line(object):
    # a=0
    # b=0
    # c=0
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2


def GetLinePara(line):
    line.a =line.p1.y - line.p2.y
    line.b = line.p2.x - line.p1.x
    line.c = line.p1.x *line.p2.y - line.p2.x * line.p1.y


def GetCrossPoint(l1,l2):

    GetLinePara(l1)
    GetLinePara(l2)
    d = l1.a * l2.b - l2.a * l1.b
    p=Point()
    p.x = (l1.b * l2.c - l2.b * l1.c)*1.0 / d
    p.y = (l1.c * l2.a - l2.c * l1.a)*1.0 / d
    return p


a=Point()
b=Point()
c=Point()
d=Point()
a.x,a.y,b.x,b.y=map(float,input('请输入第一条线的坐标>>>').split())
c.x,c.y,d.x,d.y=map(float,input('请输入第二条线的坐标>>>').split())
p1=Point(a.x,a.y)
p2=Point(b.x,b.y)
line1=Line(p1,p2)
p3=Point(c.x,c.y)
p4=Point(d.x,d.y)
line2=Line(p3,p4)
Pc = GetCrossPoint(line1,line2)
print("交点坐标为:", Pc.x, Pc.y)
```
